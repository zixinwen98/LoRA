```python
import gym
import numpy as np

MAX_ITERATIONS = 500
ENV = "FrozenLake-v1"

def Q_eps_strat(s, Q, eps):
    best_a = np.argmax(Q[s])
    prob_strat = eps * np.ones_like(Q[s]) / Q[s].shape[0]
    prob_strat[best_a] = 1 - eps
    a = np.argmax(np.random.multinomial(1, prob_strat))
    return a

def qlearning(env=ENV, epsilon=0.7, alpha=0.8, gamma=0.9):
    env = gym.make(env, is_slippery=False)
    na = env.nA
    ns = env.nS
    Q = np.zeros((ns, na))
    s = env.reset()
    for i in range(MAX_ITERATIONS):
        a = Q_eps_strat(s,Q,epsilon)
        new_s, r, done, info = env.step(a)
        Q[s,a] = (1 - alpha) * Q[s,a] + alpha * (r + gamma * np.max(Q[new_s]))
        s = new_s
    return Q

def trajectory(Q,env=ENV,eps=0,gamma=0.9):
    traj = []
    env = gym.make(env, is_slippery=False)
    s = env.reset()
    for i in range(MAX_ITERATIONS):
        env.render()
        a = Q_eps_strat(s,Q,eps)
        new_s, r, done, info = env.step(a)
        traj.append(s)
        s = new_s
    traj.append(s)
    return traj


if __name__ == '__main__':
    #Q_1 = qlearning(epsilon=0)
    Q_2 = qlearning(epsilon=0.25)
    #traj_1 = trajectory(Q_1,env=ENV,eps=0)
    traj_2 = trajectory(Q_2,env=ENV,eps=0.25)
```
